```python
import random
from openpyxl import Workbook

# Function to generate random marks for a student
def generate_random_marks():
    return [random.randint(0, 100) for _ in range(10)]

# Generate student names
student_names = [f"Student_{i+1}" for i in range(60)]

# Generate marks for each student
marks = [generate_random_marks() for _ in range(60)]

# Create a workbook and select the active worksheet
wb = Workbook()
ws = wb.active

# Write headers
ws.append(['Name'] + [f'Subject_{i+1}' for i in range(10)])

# Write student names and marks to the worksheet
for name, mark_row in zip(student_names, marks):
    ws.append([name] + mark_row)

# Save the workbook
wb.save("student_marks.xlsx")
```

```python
import pandas as pd
import numpy as np

df = pd.read_excel(r"C:\Users\pawan\OneDrive\Desktop\sample.xlsx")

l = list(df) 
a = np.array(df)

d = {}

for i in range(60):
    stu = {}
    name = a[i][1].lower()
    stu['roll no'] = a[i][0]
    stu['name'] = a[i][1]
    stu['nepali'] = a[i][2]
    stu['english'] = a[i][3]
    stu['math'] = a[i][4]
    stu['account'] = a[i][5]
    stu['science'] = a[i][6]
    stu['social'] = a[i][7]
    stu['history'] = a[i][8]
    stu['geography'] = a[i][9]
    stu['health'] = a[i][10]
    stu['sports'] = a[i][11]
    stu['attendance'] = a[i][12]
    stu['G.R. No.'] = a[i][13]
    stu['Total marks'] = a[i][14]
    stu['rank'] = a[i][15]
    d[name] = stu


nam = input("Enhter the student name - ").lower()

for i in range(60):
    if nam == a[i][1].lower():
        print("\n......generating the report card ......")
        break
else:
    print("The name is incorrect. Please check and try again.")

def calculate_grades(k):
    if k >= 91 and k <= 100:
        return "O"
    elif k >= 81 and k <= 90:
        return "A+"
    elif k >= 71 and k <= 80:
        return "A"
    elif k >= 61 and k <= 70:
        return "B+"
    elif k >= 56 and k <= 60:
        return "B"
    elif k >= 51 and k <= 55:
        return "C"
    elif k >= 33 and k <= 50:
        return ("P")
    else:
        return "N.I."

def Get_remarks(k):
    if k == 'O':
        return 'Outstanding'
    if k == 'A+':
        return 'Excellent'
    if k =='A':
        return 'Very Good'
    if k == 'B+':
        return 'Good'
    if k == 'B':
        return 'Above Average'
    if k == 'C':
        return 'Average'
    if k == 'P':
        return 'Pass'
    if k == 'N.I.':
        return 'Needs Improvement'
        
li = list(d[nam])
di = dict(d[nam])

grade = []
for i in range(2,12):
    g = calculate_grades(di[li[i]])
    grade.append(g)

remarks = []
for i in grade:
    r = Get_remarks(i)
    remarks.append(r)


def get_grades(k):
    print(f"\n\nName -- {k.upper()}")
    print(f"\n\nSubject\t\tTotal Marks\tMarks Obtained\t\tGrade\tRemarks\n")
    print(f"Nepali\t\t100\t\t\t{d[k]['nepali']}\t\t{grade[0]}\t{remarks[0]}")
    print(f"English\t\t100\t\t\t{d[k]['english']}\t\t{grade[1]}\t{remarks[1]}")
    print(f"Mathematics\t100\t\t\t{d[k]['math']}\t\t{grade[2]}\t{remarks[2]}")
    print(f"Account\t\t100\t\t\t{d[k]['account']}\t\t{grade[3]}\t{remarks[3]}")
    print(f"Science\t\t100\t\t\t{d[k]['science']}\t\t{grade[4]}\t{remarks[4]}")
    print(f"Social\t\t100\t\t\t{d[k]['social']}\t\t{grade[5]}\t{remarks[5]}")
    print(f"History\t\t100\t\t\t{d[k]['history']}\t\t{grade[6]}\t{remarks[6]}")
    print(f"Geography\t100\t\t\t{d[k]['geography']}\t\t{grade[7]}\t{remarks[7]}")
    print(f"Health\t\t100\t\t\t{d[k]['health']}\t\t{grade[8]}\t{remarks[8]}")
    print(f"Sports\t\t100\t\t\t{d[k]['sports']}\t\t{grade[9]}\t{remarks[9]}")
    Total = d[k]['nepali'] + d[k]['english'] + d[k]['math'] + d[k]['science'] + d[k]['account'] + d[k]['social'] + d[k]['history'] + d[k]['geography'] + d[k]['health'] + d[k]['sports']                       
    print(f"\n\tTotal   ------  {Total}")
    p = (Total/1000)*100
    print(f"\tPercentage ---- {p}%")
    

#    Min	Max	Grade	Remark	%
#    91	100	O	Outstanding	100
#    81	90	A+	Excellent	90
#   71	80	A	Very Good	80
#   61	70	B+	Good	70
#   56	60	B	Above Average	60
#  51	55	C	Average	55
#  33	50	P	Pass	50
#   0	32	N.I.	Needs Improvement	32

Total = d[nam]['nepali'] + d[nam]['english'] + d[nam]['math'] + d[nam]['science'] + d[nam]['account'] + d[nam]['social'] + d[nam]['history'] + d[nam]['geography'] + d[nam]['health'] + d[nam]['sports']                       
p = (Total/1000)*100

get_grades(nam)
overall_grade = calculate_grades(p)

print(f"\tOverall Grade - {overall_grade} ")
print(f"\tRank ------     {d[nam]['rank']}")
```
