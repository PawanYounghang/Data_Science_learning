```python
#concat
import pandas as pd

data1 = {'name' : ['ram','shyam'],'age':[23,34]}
df1 = pd.DataFrame(data1)

data2 = {'name' : ['Gita','Hari'],'age':[43,56]}
df2 = pd.DataFrame(data2)

df3 = pd.concat([df1,df2])
df3
df3['gender'] = ['M','M','M','F']
df3.to_csv('list.csv')
df3.to_xml('listx.xml')
df3.to_json('list.json')
```
